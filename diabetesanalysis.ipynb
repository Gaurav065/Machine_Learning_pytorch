```python
import torch
import pandas as pd
import numpy as np
from torch.utils.data import Dataset, DataLoader, random_split
import math
import torch.nn as nn
from sklearn.preprocessing import StandardScaler
```

```python
data = pd.read_csv('D:\Coding\Python Programming\Modules\Pytorch\csvs\diabetes.csv')
```

```python
inputs = data[['Pregnancies', 'Glucose', 'BloodPressure', 'SkinThickness', 'Insulin', 'BMI', 'DiabetesPedigreeFunction', 'Age']]
targets = data[['Outcome']]
```

```python
class DiabetesDataset(Dataset):
    def __init__(self):
        data = np.loadtxt('D:\Coding\Python Programming\Modules\Pytorch\csvs\diabetes.csv', delimiter=',', dtype=np.float32, skiprows= 1)
        sc = StandardScaler()
        
        x = sc.fit_transform((data[:,:8]))
        self.x = torch.from_numpy(x)
        self.y = torch.from_numpy(data[:,[8]])
        self.n_samples = data.shape[0]
    
    def __getitem__(self, index):
        return self.x[index], self.y[index]
    
    def __len__(self):
        return self.n_samples
dataset = DiabetesDataset()

batch_size = 20
train_data, test_data = random_split(dataset, [600,168])

train_dl = DataLoader(dataset=train_data, shuffle = True, batch_size= batch_size)
test_dl = DataLoader(dataset= test_data, batch_size= batch_size)
dataiter = iter(train_dl)
data = dataiter.next()
```

```python
num_epochs = 2
total_samples = len(train_data)
n_iteration = math.ceil(total_samples/20)
```

```python
for epoch in range(num_epochs):
    for i,(x_train,y_train) in enumerate(train_dl):
        if (i+1) % 10 == 0:    
            print("epoch: {}/{}, step: {}/{}, inputs: {}".format(epoch+1, num_epochs, i+1, n_iteration, x_train.shape))
```

```python

class DiabetesModel(nn.Module):
    def __init__(self, input_size, layer_size, output_size):
        super().__init__()
        self.linear = nn.Linear(input_size,layer_size)
        self.linear2 = nn.Linear(layer_size,output_size)
    
    def forward(self, x):
        preds = self.linear(x)
        preds = torch.relu(preds)
        preds = self.linear2(preds)
        preds = torch.sigmoid(preds)
        return preds

model = DiabetesModel(8,4,1)
loss_fn = nn.BCELoss()
optimizer = torch.optim.SGD(model.parameters(), lr = 0.001)

        
```

```python
epochs = 1000

for epoch in range(epochs):
    for i, (x_train, y_train) in enumerate(train_dl):   
        preds = model(x_train)
        loss = loss_fn(preds,y_train)
        loss.backward()
        optimizer.step()
        optimizer.zero_grad()
    # print(preds.round(), y_train)
    
    with torch.no_grad():
        correct = 0
        for i, (x_test, y_test) in enumerate(test_dl):
            preds = model(x_test)
            correct += torch.sum(preds.round()==y_test).item()
        
        accuracy = correct/168
        
print("Accuracy: ",round(accuracy*100,1),"%")
        
```
